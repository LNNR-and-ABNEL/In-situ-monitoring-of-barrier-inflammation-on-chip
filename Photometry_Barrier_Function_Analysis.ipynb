# Automated Image Analysis for Fiber Photometry

```python
import time
import pandas as pd
import numpy as np

#Importing the files as Apical and Basal folders to analyze them
folder = input('Data Folder File Path (e.g. C:/Users/.../Desktop/Data Folder):')
dirNameA = f'{folder}/Apical/*.tiff' 
dirNameB = f'{folder}/Basal/*.tiff'

record_rate = float(input('Recording rate in minutes per image:'))
EXPECTED_FIBERS = int(input('Number of expected fiber ends per channel:'))
    
%store folder
%store EXPECTED_FIBERS

#make a list here that has the string labels for each of the column
headers = ['Time (hr)'] #gathering a list of legend labels, always have time as the 0 column
for i in ['DF/F','Intensity']: #we have apical and basal fiber groups
    for j in range(EXPECTED_FIBERS): #count for number of fibers used in each group
        headers.append(f'%s {j+1}' %i) 

blank_excel = pd.DataFrame(index = None, columns = headers)
with pd.ExcelWriter(f'{folder}.xlsx', engine='openpyxl') as writer:
    #INSERT THE HEADERS HERE USE LIST SLICES WITH EXPECTED_FIBERS FOR APICAL/BASAL SPLIT
    blank_excel.to_excel(writer, sheet_name='Apical', index = None)
    blank_excel.to_excel(writer, sheet_name='Basal', index = None)
```
